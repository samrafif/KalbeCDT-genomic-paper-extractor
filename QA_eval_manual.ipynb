```python
import os
from dotenv import load_dotenv

import gradio as gr
from langchain_groq import ChatGroq

from extractor import Answerer, Store
```

```python
load_dotenv()

vec_store = Store("main", doc_k=5)
vec_store.setup()
```

```python
llm = ChatGroq(
    model_name="llama-3.1-8b-instant",
    temperature=0.3
)

q_answerer = Answerer(vec_store, model=llm)
```

```python
results = []
```

```python
query = "Genotypes of the APOE gene and their traits"
result, new_context_docs, cits = q_answerer.answer_with_search(query)
results.append(result[-1]['content'])
```
